<b>Ans2.</b><br>
In Ricart and Agrawalaâ€™s multicast based mutual exclusion algorithm, a client issues
a multicast request every time it requires entry. This is inefficient in the case

outstanding requests could mark the token as JUST RELEASED , meaning that it has

critical section.<br>
To meet liveness condition ME2, a JUST RELEASED token should become RELEASED
if a request for entry is received.

<h3>The Atomic Commit Protocol(ACP) Problem</h3>
<li>Validity</li>If a coordinator broadcasts a message m, then all participants eventually receive m
<li>Integrity</li>For any message m, each participant receives m at most once and only if a coordinator actually broadcasts m
<li>Timeliness</li>There is a known constant d (delay) such that a broadcast of m initiated at time t, is received by every participant by t + d
<h3>One-Phase Commit(1PC) Protocol</h3>
<ol>
    <li>Coordinator receives request from client</li>
    <li>Coordinator broadcasts message indicating commit or abort</li>
    <li>When all participants have sent back acknowledgment exit, otherwise broadcast instruction again</li>
</ol>
<b>Limitation</b><br>
Coordinator cannot decide to abort when client requests commit
<li>Concurrency control requires abort (deadlock, failed validation)</li>
<li>Server may have crashed and been restarted (which conveys to abort the transaction)</li>
<b>Failure Model</b><br>
<li>Process Crashes</li>
--Any process can crash at any time.
--This includes the coordinator of the transaction.
<li>Bounded Delays in Message Delivery</li>
--Messages are delayed up to d time units
--A message sent at time t will be received some time before t + d.
<h3>Two-phase Commit(2PC) Protocol</h3>
The transaction will commit only if all participants vote to do so.<br>
If commit is the decision reached, Tentative versions of shared objects are written to permanent storage. Otherwise, these changes are discarded.<br>

<h3>Three-Phase Commit(3PC) Protocol</h3>
<li>preCommit(t): Coordinator ! Participant</li>Coordinator tells participant it is prepared to commit
<li>ackPreCommit(t; p): Participant ! Coordinator</li>Acknowledges receipt of preCommit(t)
<b>Failure Handling</b><br>
<li>If participant crashes before sending ackPrecommit(t)
<ul><li>coordinator may still decide to commit transaction,</li>
<li>if so, process replacing crashed participant then completes its part of
the transaction.</li>
</ul></li>
<li>If coordinator crashes in the precommit phase,
<ul><li>participants can agree among themselves whether to commit the
transaction,</li>
<li>since they didn't roll back or make visible their tentative versions.</li>
</ul></li>
