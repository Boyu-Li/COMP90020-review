<b>Ans9.</b><br>

</ul>

<h3>The Bully-Algorithm</h3>
<b>Assunption</b><br>
<li>All processes have a unique integer ID</li>
<li>Every node can communicate with every other node</li>
<b>Message types</b>
<li>Election</li>
<li>Answer</li>
<li>Coordinator</li>
<b>Coordinator</b><br>
<li>Process decides whether to become coordinator by comparing own ID
with all other IDs (highest wins)
<li>Announce its role by sending coordinator message to all other nodes
with lower ID
<li>Process with lower ID can bid to become coordinator by sending
election message to all processes with higher ID
<li>If no response within T, considers itself elected coordinator, sends

<li>Otherwise: wait for another Tâ€™ time units for a coordinator message

<li>If no response, then begin another election process

<b>Receipt of an election message</b><br>
<li>Process sets variable electioni to the ID of the coordinator received in the election message</li>
<li>Process sends back an answer message and begins another election â€“ unless an election was already initiated</li>
<b>New process replacing crashed process</b><br>
