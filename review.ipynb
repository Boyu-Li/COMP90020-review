<h3>Clocks</h3>
<li><b>Hardware clock:</b>$H_i(t)$</li>
<li><b>Software clock:</b>$C_i(t)=\alpha H_i(t)+\beta$</li>
<h3>Skew(drift)</h3>
<li>Clocks can also drift even if they were set perfectly to be the same at the beginning</li>
<li>The speed of an individual clock can also change, e.g., with temperature</li>
<h3>Coordinated Universal Time(UTC)</h3>
UTC is broadcasted all over the world to synchronize their time (with some minor error in this sync process)
<h3>External synchronization</h3>
用权威的外部时间源同步进程的时钟<br>
<li>Use an external time server to synchronize process times</li>
<li>For a synchronization bound D > 0 and for a source S of UTC time: $|S(t) - C_i(t)| < D$ for $i = 1, 2, ..., N$ and for all real times $t$ in interval $i$</li>
<li>The clocks $C_i$ are accurate to within the bound D</li>
<h3>Internal synchronization</h3>
通过本地时钟度量在不同计算机上发生的两个事件的间隔（即使他们没有必要与外部时间同步）<br>
<li>For a synchronization bound $D > 0: |Ci(t) - Cj(t)| < D$ for $i = 1, 2, ..., N$ and for all real times $t$ in interval $i$</li>
<li>The clocks $C_i$ agree within the bound D</li>
<li>Once synchronized, processes can communicate</li>
<li>If all clocks drift then all can become different than the initial external time service</li>
<h3>Monotonicity condition(单调性条件)</h3>
A clock always advances: $t^\prime > t \Rightarrow C(t^\prime)>C(t)$
<h3>Faulty Clocks(故障时钟)</h3>
<li>A clock that do not obey the monotonicity condition</li>
<li>and/or the bounds on its drift</li>
<b>A clock is said to had a crash failure(崩溃故障) if it totally stops running, else it is said to have an arbitrary failure(随即故障)</b><br>
<b>A correct clock is not necessarily an accurate clock(始终正确不一定准确)</b><br>
<h3>Cristian's Method: Asychronous System</h3>
<b>A time server S receives signals from a UTC source</b><br>
$time = t+T_{Round}/2$<br>
$Accuracy\pm (T_{round}/2 - min)$<br>
time message back: $[t+min,t+T_{round}-min]$<br>
<h3>Shortage in Cristian's algorithm</h3>
<li>A single time server might fail, need to use of a group of synchronized servers</li>
<li>It does not deal with faulty servers</li>
<h3>Berkeley Algorithm</h3>
<li>An algorithm for internal synchronization of a group of computers</li>
<li>A master polls to collect clock values from the others (slaves)</li>
<li>The master uses round trip times to estimate the slaves’ clock values</li>
<li>It takes an average (eliminating any above some average round trip time or with faulty clocks)</li>
<li>It sends the required adjustment to the slaves (why not the updated times?)</li>
<h3>Network Time Protocol(NTP)</h3>
<li>Primary servers are connected to UTC</li>
<li>Secondary servers are synchronized to primary servers</li>
<li>Synchronization subnet - lowest level servers in users' computers</li>
A time service for the Internet - synchronizes clients to UTC<br>
Reliability from redundant paths, scalable, authenticates time sources<br>
<h3>Modes of synchronization</h3>
<li>Multicast</li>
--A server within a high speed LAN multicasts time to others which set clocks assuming some delay (not very accurate)
<li>Procedure call</li>
--A server accepts requests from other computers (like Cristian’s algorithm). Higher accuracy. Useful if no hardware multicast.
<li>Symmetric</li>
--Pairs of servers exchange messages containing time information<br>
--Used where very high accuracies are needed (e.g. for higher levels)

<h3>Happened-Before Relation(HB)</h3>
<li>HB1:If $\exists$ process $p$: $a \rightarrow_p b$, then $a \rightarrow  b$</li>
<li>HB2:For any message $m$: a = send($m$) $\rightarrow b$ = receive($m$)</li>
<li>HB3:If $a \rightarrow b$ and $b \rightarrow c$, then $a \rightarrow c$</li>
<b>a and e are concurrent; write as a || e</b><br>
<h3>Lamport's Logical Clocks</h3>
<li>LC1:</li>
--$L_i$ is incremented by 1 before each event at process $p_i$</br>
<li>LC2:</li>
--When process $p_i$ sends message $m$, it piggybacks $t = L_i$<br>
--When $p_j$ receives $(m,t)$ it sets $L_j$ := $max(L_j, t)$ and applies LC1 before time stamping the event $receive(m)$<br>
<b>Important</b><br>
<li>Each process has its logical clock initialized to zero</li>
<li>$e \rightarrow e^\prime$ implies $L(e) < L(e^\prime)$</li>
<li>However, $L(e) < L(e^\prime)$ does not imply $e\rightarrow e^\prime$</li>
<h3>Vector Clocks</h3>
<li>Vector clock Vi at process pi is an array of N integers</li>
<li>VC1: set $V_i[j]$ = 0 for $i, j = 1, 2, …, N$</li>
<li>VC2: before $p_i$ timestamps an event it sets $V_i[i] := V_i[i] + 1$</li>
<li>VC3: $p_i$ piggybacks $t = V_i$ on every message it sends</li>
<li>VC4: when $p_i$ receives $(m, t)$ it sets $V_i[j] := max(V_i[j], t[j]), j = 1, 2,…, N$(and adds 1 before the next event to its own element using VC2)</li>
<h3>Vector Clocks Comparison</h3>
<li>$V = V^\prime$ iff $V [j] = V^\prime [j] $for$ j = 1, 2, ..., N$</li>
<li>$V <= V^\prime$ iff $V [j] <=V^\prime [j] $for$ j = 1, 2, ..., N$</li>
<li>$V < V^\prime$ iff $V <= V^\prime $and$ V != V^\prime$</li>
<li>$e \rightarrow e^\prime$ iff $V(e) < V(e^\prime)$</li>

<h3>Q9</h3>
<b>Discuss how it is possible to compensate for clock drift
between synchronization points by observing the drift
rate over time. Discuss any limitations to your method.</b><br>
<b>Ans9.</b><br>
If we know that the drift rate is constant, then we need only measure it between synchronization points with an accurate source and compensate for it. For example, if the clock loses a second every hour, then we can add a second every hour, in smooth increments, to the value returned to the user. The difficulty is that the clock’s drift rate is liable to be variable – for example, it may be a function of temperature. Therefore we need an adaptive adjustment method, which guesses the drift rate, based on past behaviour, but which compensates when the drift rate is discovered to have changed by the next synchronisation point.

<h3>Q11</h3>

<h2>Coordination and Agreement</h2>

<h3>Assumptions</h3>
<li>Processes are connected via reliable channels</li>
<li>Processes do not rely on others to communicate</li>
<li>Processes themselves can crash or act unexpectedly</li>
<h3>Conditions for Mutual Exclusion</h3>
<li>ME1:(Safety)</li>临界区一次最多有一个进程可以执行。
<li>ME2:(Liveness)</li>进入和离开临界区的请求最终执行成功。(无死锁和饥饿)
<li>ME3:($\rightarrow$Ordering)</li>如果一个进入CS的请求发生在先，那么进入CS时仍按此顺序。
<h3>Central Server Algorithm</h3>
<li>Central server receives access requests
<ul>
<li>If no process in critical section, request
will be granted
<li>If process in critical section, request will
be queued
</ul>
    </li>
<li>Process leaves critical section</li>
<li>Grant access to next process in queue or wait for new requests if queue is empty</li>
<b>Safety and Liveness are guaranteed</b><br>
<b>Ordering (ME3) is not guaranteed(Network delays may reorder requests)</b><br>
<b>Performance and availability of server are the bottlenecks</b><br>

<h3>Evaluation of Mutual Exclusion ALgorithms</h3>
<li>Bandwidth consumption</li>在每个entry和exit操作中发送的消息数
<li>Client Delays</li>在每一次entry和exit操作由进程导致的客户延迟
<li>Throughput</li>系统允许的CS访问次数<br>
---• Measure effect using the synchronization delay of one process exiting
the critical section and the next process entering it<br>
---• Throughput is greater when the synchronization delay is shorter

<h3>Ring-based Algorithm</h3>
<li>Token arrival
    <ul>
<li>Only process in possession of token
may access critical region</li>
<li>If no request upon arrival of token,
or when exiting critical region, pass
token on to neighbor</li>
    </ul>
</li>
<li>Algorithm satisfies ME1, ME2, but not ME3(happened before relation)</li>
<li>Performance
<ul>
<li>Constant bandwidth consumption</li>
<li>Entry delay between 0 and N
message transmission times</li>
<li>Synchronization delay between 1
and N message transmission times</li>
<li>The algorithm wastes bandwidth if
no process requests the token</li>
</ul>
</li>
<h3>Using Multicast and Logical Clocks(Ricart and Agrawala's Algorithm)</h3>
<li>Each process requests to enter a critical section via multicast</li>
<li>When others reply to the multicast, then you can decide on the request</li>
<li>All the conditions ME1 – ME3 are met</li>
<li>Processes use Lamport clocks and their IDs
<ul>
<li>Timestamps are used to resolve simultaneous requests</li>
<li>If two timestamps are the same, then process IDs are used</li>
</ul>
</li>
<img src="./img/7.png\"/><br>
<b>Performance</b><br>
<li>Expensive access requires 2(N-1) messages per request</li>
<li>Synchronization delay: just one message transmission time (previous algorithms: up to N)</li>
<li>Improved versions exist for reducing bandwidth costs, e.g., repeated entry of same process without executing protocol</li>
<b>IMPROVEMENT</b><br>
针对最近一次进入过临界区且没有接收到其他的进入请求的进程，见tutorial
<h3>Maekawa's Voting Algorithm</h3>
<img src="./img/8.png\"/><br>
<b>Satisfies ME1</b><br>
$V_i\cap V_j\neq \phi$中的进程必须对两个中的一个投票<br>
<b>Deadlocks are possible</b><br>$V_1={p_1,P_2},V_2={p_2,P_3},V_3={p_3,P_1}$<br>
每个进程收到两个中的一个，都不能继续，不满足ME2
<h3>Extensions for ME2 and ME3</h3>
<li>Modificaion to ensure absence of deadlocks</li>
<li>Use of logical clocks:processes queue requests in happened-before order</li>
<h3>Performance</h3>
$2(N-1)$ versus. $3\sqrt{N}$($\sqrt{N}$ for request response and release)<br>
Better than Ricart and Agrawala for N>4

<h3>Failures容错</h3>
<li>All of the covered algorithms cannot tolerate a crash failure(Ring-algorithms cannot tolerate a single crash failure)</li>
<li>Maekawa 算法可以容忍一些进程的崩溃故障：如果一个崩溃进程不在所需的投票集中，那么它的故障不会影响其他进程</li>
<li>中央服务器算法可以容忍一个既不请求也不持有令牌的客户进程的崩溃故障。</li>
<li>可通过隐式地给所有请求授权来修改Ricart and Agrawala算法，使得它容忍进程地崩溃故障</li>
<h3>Elections requirements($\bot$:a special value means undefined)</h3>
<li>E1:a process $p_i$ has $elected_i = \bot(undefined) or elected_i = P$ for some
non-crashed process $P$ that will be chosen at the end of the run with
the largest identifier (safety)</li>
<li>E2:all processes $p_i$ will eventually set $elected_i \neq\bot(liveness)$ or crash</li>

<h3>Ring-based Election Algorithm</h3>
<b>Assumptions</b><br>
<li>All processes have a unique integer ID</li>
<li>All N nodes communicate on a uni-directional ring structure</li>
<li>Each process $p_i$ has a communication channel to the next process in
the ring, $p_{(i+1) mod N}$</li>
<b>Algorithm</b><br>
<li>Initially all processes marked as non-participants</li>
<li>Any process can start an election</li>
<li>Election start: a process marks itself as a participant and places its ID
in a message and sends it clockwise to the next</li>
<b>Ring-based Election</b><br>
<ul>
<li>If a message is received, a process compares the received ID with its own
<ul>
<li>If the ID is greater then forward it to the next process
<li>If the ID is smaller then
<ul>
<li>If own status is “non-participant”, then substitute own ID in election message
and forward on ring
<li>otherwise, do not forward message (as “participant”)
</ul>
<li>If received ID is identical to own ID
<ul>
<li>This process ID must be greatest and it becomes elected
<li>Marks own status as “non-participant”
<li>Sends out “elected” message
</ul>
</ul>
<li>If a message is forwarded, upon any forwarding, mark own state as “participant”
<li>When receiving “elected” message
<ul>
<li>Mark own status as “non-participant”
<li>Sets its variable electedi to the ID in the message (unless it is the new
coordinator)
<li>Forward elected message
</ul>
</ul>
<b>Performance</b><br>
<li>E1:Safety</li>Met as all identifiers are considered. Even if elections start simultaneously the smaller identifier cannot pass the larger identifier
<li>E2:Liveness</li>Met due to reliable communication
<li>Worst-case behavior</li>
3N – 1 messages in total and sequentially
<li>Failures</li>
Not tolerated
<h3>The Bully-Algorithm</h3>
<b>Assunption</b><br>
<li>All processes have a unique integer ID</li>
<li>Every node can communicate with every other node</li>
<b>Message types</b>
<li>Election</li>
<li>Answer</li>
<li>Coordinator</li>
<b>Coordinator</b><br>
<li>Process decides whether to become coordinator by comparing own ID
with all other IDs (highest wins)
<li>Announce its role by sending coordinator message to all other nodes
with lower ID
<li>Process with lower ID can bid to become coordinator by sending
election message to all processes with higher ID
<li>If no response within T, considers itself elected coordinator, sends
coordinator message to all processes with lower ID
<li>Otherwise: wait for another T’ time units for a coordinator message
to arrive from new coordinator
<li>If no response, then begin another election process

<b>Receipt of an election message</b><br>
<li>Process sets variable electioni to the ID of the coordinator received in the election message</li>
<li>Process sends back an answer message and begins another election – unless an election was already initiated</li>
<b>New process replacing crashed process</b><br>
<li>If it has the highest ID, it will immediately send coordinator message and “bully” current coordinator to resign</li>

<b>The Bully Algorithm: Evaluation</b><br>
<li>Properties
    <ul>
        <li>E1 satisfied if no process is replaced and the timeout T estimate is accurate</li>
        <li>E2 satisfied for a synchronous network and reliable transmission</li>
        <li>E1 not satisfied if crashed process replaced at the same time while another process has announced that it is the new coordinator</li>
    </ul>
</li>
<li>Performance
    <ul>
        <li>Best case: process with the second highest identifier detects coordinators failure and elects itself coordinator and sends N-2 coordinator messages</li>
        <li>requires $O(N^2)$ messages in worst case when lowest ID detects failure
$\Rightarrow N-1$ processes with higher IDs start election</li>
    </ul>
</li>
